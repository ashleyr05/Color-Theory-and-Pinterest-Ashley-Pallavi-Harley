```python
import numpy as np

all_pixels = []

Y_black_threshold = 0.1 
Y_white_threshold = 0.9 
UV_neutral_threshold = 0.05  

for index, row in my_db.iterrows():
    try:
        img = row["image"]
        img_yuv = img.convert("YCbCr")  
        yuv_data = np.array(img_yuv)   
        y = yuv_data[:, :, 0] 
        u = yuv_data[:, :, 1] 
        v = yuv_data[:, :, 2] 

        yuv_flat = np.column_stack([y.flatten(), u.flatten(), v.flatten()])

       
        is_black = yuv_flat[:, 0] < Y_black_threshold
        is_white = (yuv_flat[:, 0] > Y_white_threshold) & \
                   (np.abs(yuv_flat[:, 1]) < UV_neutral_threshold) & \
                   (np.abs(yuv_flat[:, 2]) < UV_neutral_threshold)
        valid_pixels = ~(is_black | is_white)

        all_pixels.append(yuv_flat[valid_pixels])
    except Exception as e:
        print(f"Failed to process image at {row['url']}: {e}")

if all_pixels:
    all_pixels_array = np.vstack(all_pixels)

    overall_avg_color = np.mean(all_pixels_array, axis=0)
    overall_std_dev = np.std(all_pixels_array, axis=0)

    print("Overall Average Color (YUV):", overall_avg_color)
    print("Overall Standard Deviation (YUV):", overall_std_dev)
else:
    print("No valid image data was found.")


#Use https://convertingcolors.com/yuv-color-71.2000_10.9500_14.4000.html?search=YUV(71.2,%2010.95,%2014.4) to interpret result
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = []

Y_black_threshold = 0.1
Y_white_threshold = 0.9
UV_neutral_threshold = 0.05
resizing = (100, 100)

for index, row in my_db.iterrows():
    try:
        img = row["image"]
        img_resized = img.resize(resizing)
        img_yuv = img.convert("YCbCr")
        yuv_data = np.array(img_yuv)

        y = yuv_data[:, :, 0] / 255.0
        u = (yuv_data[:, :, 1] - 128) / 255.0
        v = (yuv_data[:, :, 2] - 128) / 255.0

        yuv_flat = np.column_stack([y.flatten(), u.flatten(), v.flatten()])

        is_black = yuv_flat[:, 0] < Y_black_threshold
        is_white = (yuv_flat[:, 0] > Y_white_threshold) & \
                   (np.abs(yuv_flat[:, 1]) < UV_neutral_threshold) & \
                   (np.abs(yuv_flat[:, 2]) < UV_neutral_threshold)

        valid_pixels = ~(is_black | is_white)

        a.append(yuv_flat[valid_pixels])
    except Exception as e:
        print(f"")

if a:
    ar = np.vstack(a)

    yuv_palette = {}
    for i, channel in enumerate(['Y', 'U', 'V']):
        channel_data = ar[:, i]
        q1 = np.percentile(channel_data, 25)
        q2 = np.median(channel_data)
        q3 = np.percentile(channel_data, 75)
        min_val = channel_data.min()
        max_val = channel_data.max()

        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = channel_data[(channel_data < lower_bound) | (channel_data > upper_bound)]

        yuv_palette[channel] = {
            "min": min_val,
            "Q1": q1,
            "median": q2,
            "Q3": q3,
            "max": max_val,
            "outliers": outliers
        }

    for channel, stats in yuv_palette.items():
        print(f"{channel} Channel:")
        print(f"  Min: {stats['min']}")
        print(f"  Q1: {stats['Q1']}")
        print(f"  Median: {stats['median']}")
        print(f"  Q3: {stats['Q3']}")
        print(f"  Max: {stats['max']}")
        print(f"  Outliers: {stats['outliers'][:10]}...")

    plt.boxplot([ar[:, 0], ar[:, 1], ar[:, 2]],
            labels=['Y', 'U', 'V'], 
            patch_artist=True,
            medianprops=dict(color='white', linewidth=2))
    plt.title("YUV Color Distribution-- Warm Tone")
    plt.ylabel("Normalized Value")
    plt.grid(True)
    plt.show()

else:
    print("No valid image data was found.")
```
