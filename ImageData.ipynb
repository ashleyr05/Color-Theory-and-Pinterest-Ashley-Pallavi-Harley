```python
from PIL import Image
import requests
from io import BytesIO
import pandas as pd

data = {"url":[], "image":[]}
my_db = pd.DataFrame(data=data)
url = 'https://i.pinimg.com/150x150/0d/f6/f1/0df6f1f0bfe7aaca849c1bbc3607a34b.jpg'
url2 = 'https://i.pinimg.com/474x/0e/f8/e8/0ef8e87826d3bb709c8ae0925323105f.jpg'
url_list = [url,url2]
for url in url_list:
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    step_data = {'url':[url],'image':img}
    step_db = pd.DataFrame(data=step_data)
    my_db=pd.concat([my_db, step_db])
```
